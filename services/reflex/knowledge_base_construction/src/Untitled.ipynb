```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import pickle
```

```python
with open('viz.pkl', 'rb') as rf:
    data = pickle.load(rf)
    
print(data)
```

```python
fig, axes = plt.subplots(nrows=len(data))
fig.set_size_inches(100, 40)
for ind, seq in enumerate(data):
    words, values = zip(*seq)
    words = np.expand_dims(np.array(words), axis=0)
    values = np.expand_dims(np.array(values), axis=0)
    sb.heatmap(values, ax=axes[ind], annot=words, fmt='')
plt.savefig('test7.png')
```
